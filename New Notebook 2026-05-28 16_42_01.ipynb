```python
df = spark.read.option("header", True).option("inferSchema", True).csv("/Volumes/workspace/default/agent_etl/Agn.csv")
display(df.limit(5))
df.createOrReplaceTempView("Agent_elt_sql")
```

```python
%sql
select incident_id, count(*) as total_count
from Agent_elt_sql
group by incident_id
order by total_count desc
limit 4
```

```python
%sql
select incident_first_solved_agent_id,
count(incident_id) as total_incidents,
sum(CASE WHEN resolution_re_deadline_breached = 'Breach' then 1 else 0 end) as total_breach,
round(avg(pain_in_mins),2) as  avg_customer_pain,
round(sum(CASE WHEN resolution_re_deadline_breached = 'Breach' then 1 else 0 end) *100/count(incident_id),2) as pct_change
from Agent_elt_sql
where incident_first_solved_agent_id is not null
group by incident_first_solved_agent_id
order by total_breach desc, avg_customer_pain DESC
limit 10
```

```python
%sql
select
courier_partner,
count(incident_id) as total_delivery_issues,
sum(case when resolution_deadline_breached = 'Breach' then 1 else 0 end) as sla_breach,
round(sum(case when resolution_deadline_breached = 'Breach' then 1 else 0 end) / count(incident_id),2) as pct_change
from Agent_elt_sql
group by courier_partner
order by  total_delivery_issues desc, pct_change desc
limit 10
    
```

```python
%sql
select 
incident_id,
customer_id,
pain_in_mins,
escalation_flag,
final_issue_type,
status_name
from Agent_elt_sql
where pain_in_mins > 180 and escalation_flag <> TRUE
order by pain_in_mins desc
limit 5
```

```python
%sql
select
final_issue_type,
count(incident_id) as issue_count,
sum(pain_in_mins) as total_pain,
round(avg(pain_in_mins),2) as avg_pain_ticket,
round(avg(date_diff(TO_TIMESTAMP(assigned_to_queue_at,'dd-MM-yyyy HH:mm'), to_timestamp( incident_creation_time,'dd-MM-yyyy HH:mm'))),2) as date_diff_days
from Agent_elt_sql
group by final_issue_type
order by total_pain desc, date_diff_days DESC
limit 10
```

```python
%sql
select
current_queue_name,
count(incident_id) as total_count,
avg(pain_in_mins) as avg_pain,
sum(case when resolution_deadline_breached = 'Breach' then 1 else 0 end) as total_pain_count
from Agent_elt_sql
group by current_queue_name
order by  total_pain_count DESC
limit 10

```

```python
%sql
select
customer_id,
count( distinct incident_id) as unique_incident,
sum(pain_in_mins) as total_pain,
min(incident_creation_date) as first_incident,
max(incident_creation_date) as last_incident
from Agent_elt_sql
group by customer_id
having unique_incident > 1
order by unique_incident DESC, total_pain desc
limit 10

```

```python
%sql
select 
first_queue_name,
current_queue_name,
count(incident_id) as incident_count
from Agent_elt_sql
where first_queue_name <> current_queue_name
group by first_queue_name,current_queue_name
limit 5
```

```python
%sql
with queue_data as 
(select incident_id,
first_queue_name,
current_queue_name
from Agent_elt_sql
where first_queue_name <> current_queue_name
)
select 
first_queue_name,
current_queue_name,
count(*) as total_count
from queue_data
group by first_queue_name,current_queue_name
limit 10
```
